```python

def create_game_board():
    return [[0] * 10 for _ in range(10)]

def print_game_board(board):
    print("  ", end="")
    for col in range(10):
        print(col, end=" ")
    print()

    for row in range(10):
        print(row, end=" ")
        for col in range(10):
            print(board[row][col], end=" ")
        print()

def update_prompt(board):
    try:
        row = int(input("Enter a row number (0-9): "))
        col = int(input("Enter a column number (0-9): \n"))
    except ValueError:
        return False

    if 0 <= row < 10 and 0 <= col < 10:
        board[row][col] = 1
        return True
    else:
        return False

def main():
    game_board = create_game_board()
    while True:
        print_game_board(game_board)
        if not update_prompt(game_board):
            break
    print("Exiting the game.")

if __name__ == "__main__":
    main()
```
